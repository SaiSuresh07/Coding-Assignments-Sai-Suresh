```python
#Report Card Dictionary

report_card_dictionary ={"student_names":["Jay","Aryan","Sahil","Riya","Shubh"],
                        "math_scores": [20,56,76,78,90],
                         "chemistry_scores":[32,56,78,23,100],
                         "information_technology_scores":[34,36,48,100,45],
                         "physics_scores":[6,88,58,90,25] }
report_card_dictionary
```

```python
#Setup
import pandas as pd
```

```python
#Converting Dictionary to DataFrame
report_card_df = pd.DataFrame(report_card_dictionary)
report_card_df
```

```python
#First two rowsin DataFrame
report_card_df.head(2)
```

```python
#Sample of two rows in our DataFrame
report_card_df.sample(3)
```

```python
#Create a description of our numeric columns
report_card_df.describe()
```

```python
#Create a column called total score

report_card_df["total_score"] = report_card_df["math_scores"] + report_card_df["chemistry_scores"] + report_card_df["information_technology_scores"] + report_card_df["physics_scores"]
report_card_df.head(5)
```


Create a column called average score


```python
report_card_df["average_score"] = report_card_df["total_score"]/ 4
report_card_df
```

```python
#Describe the report card dataframe
report_card_df.describe()
```

# Insights

- **What is the mean chemistry score?** 62.6
- **What is the lowest physics score?** 25
- **What is the highest total score?** 291
- **What is the mean total score?** 248
- **What is the mean average score?** 62

```python
#Creating a Table called Report Card Insights
report_card_insights_df = report_card_df.describe().transpose()
report_card_insights_df
```

```python
#Creating a CSV File of our Report Card Insights

report_card_insights_df.to_csv("Report Card Insight.csv")
```

```python
#Create a column called Status that contains a value that states whether a student passed or failed the wuarter semester
report_card_df.loc[report_card_df["total_score"]>=140, "status"] = "pass"
report_card_df.loc[report_card_df["total_score"]<140, "status"] = "fail"
```

```python
report_card_df
```

```python
#Create a column called Character Grade that contains the Character Grade for the student
report_card_df.loc[report_card_df["average_score"]>=90,"character_grade"] = "A"
report_card_df.loc[(report_card_df["average_score"]>=70)&(report_card_df["average_score"]<90),"character_grade"] = "B"
report_card_df.loc[(report_card_df["average_score"]>=50)&(report_card_df["average_score"]<70),"character_grade"] = "C"
report_card_df.loc[(report_card_df["average_score"]>=35)&(report_card_df["average_score"]<50),"character_grade"] = "D"
report_card_df.loc[report_card_df["average_score"]<35,"character_grade"] = "F"
```
