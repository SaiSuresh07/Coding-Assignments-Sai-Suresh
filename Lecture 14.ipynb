```python
# Setup
import pandas as pd
```

```python
sales_2017_df = pd.read_csv("Australian_Grocery_Store_Data_2017.csv")
sales_2017_df.head(5)
```

```python
sales_2018_df = pd.read_csv("Australian_Grocery_Store_Data_2018.csv")
sales_2018_df.head(5)
```

```python
sales_2018_df.sort_values("receipt_id").head(10)
```

```python
sales_df = pd.concat((sales_2017_df, sales_2018_df), ignore_index= True)
sales_df.sort_values("receipt_id").head(10)
```

```python
# 
items_categories_df = pd.read_csv("Australian_Grocery_Store_Data_Only_Items_Categories.csv")
items_categories_df
```

```python
sales_df = pd.merge(sales_df, items_categories_df, how = "left", on = ["item_name"])
sales_df
```

```python
# coverting all string to title case
sales_df["item_name"] = sales_df["item_name"].str.title()
sales_df["payment_type"] = sales_df["payment_type"].str.title()
sales_df["main_category"] = sales_df["main_category"].str.title()
sales_df["sub-category"] = sales_df["sub_category"].str.title()
sales_df
```

## Add Columns

- Total Buying Price = unit_buying_price
- Total selling price = unit_selling_price
- Total Margin = unit_per_margin

```python
sales_df["total_buying_price"] = sales_df["quantity"] * sales_df["unit_buying_price"]
sales_df["total_selling_price"] = sales_df["quantity"] * sales_df["unit_selling_price"]
sales_df["total_margin"] = sales_df["quantity"] * sales_df["unit_price_margin"]
sales_df
```

```python
sales_df.head(5)
```

## I want u to create a colum called quantity type
- if quantity is <= 1.100 then assing the value 'single
- if quantity is >1.100 and quantity is <10 then assign the value as 'regular
- if quantity is>=10 then assign the value as 'bulk

```python
sales_df.loc[sales_df["quantity"]<=1.100, "quantity_type"] = "single"
sales_df.loc[(sales_df["quantity"]>1.100) & (sales_df["quantity"]<10), "quantity_type"] = "regular"
sales_df.loc[sales_df["quantity"]>=10, "quantity_type"] = "bulk"
sales_df.sample(10)
```

```python
# date column
sales_df.dtypes
```

```python
sales_df["date"] = pd.to_datetime(sales_df["date"])
```

```python
sales_df.dtypes
```

```python
sales_df["year"] = sales_df["date"].dt.year
sales_df["quarter"] = sales_df["date"].dt.quarter
sales_df["month"] = sales_df["date"].dt.month_name()
sales_df["day"] = sales_df["date"].dt.day_name()
sales_df["hour"] = sales_df["date"].dt.hour
sales_df.head(10)
```

# Item Analysis

```python
items_df = sales_df.groupby(["item_name", "main_category", "sub_category"]).agg(unit_buying_price = ("unit_buying_price", "mean"),
                                                                               unit_selling_price = ("unit_selling_price", "mean"),
                                                                                unit_price_margin = ("unit_price_margin", "mean"),
                                                                                quantity = ("quantity", "sum"),
                                                                                revenue = ("total_selling_price", "sum"),
                                                                                profit = ("total_margin", "sum"))
items_df
```
